# Air Passengers Trend and Seasonaility -- Time series

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import statsmodels.api as sm
import warnings
```

```python
ap = pd.read_csv('AirPassengers.csv')
```

```python
print(ap)
```

```python
ap.head()
```

```python
ap.columns = ['Month' , 'Number of Passengers']

ap.head()
```

```python
#Visualize timeseries
```

```python
def plot_df(ap, x, y, title="", xlabel='Month', ylabel='Number of Passengers', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    

plot_df(ap, x=ap['Month'], y=ap['Number of Passengers'], title='Number of US Airline passengers from 1949 to 1960')
```

```python
#both sides of the Y axis to emphasize the growth.
```

```python
x = ap['Month'].values
y1 = ap['Number of Passengers'].values

# Plot
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi= 120)
plt.fill_between(x, y1=y1, y2=-y1, alpha=0.5, linewidth=2, color='seagreen')
plt.ylim(-800, 800)
plt.title('Air Passengers (Two Side View)', fontsize=16)
plt.hlines(y=0, xmin=np.min(ap['Month']), xmax=np.max(ap['Month']), linewidth=.5)
plt.show()
```

```python
#Trend in seasonability
```

```python
def plot_df(ap, x, y, title="", xlabel='Month', ylabel='Number of Passengers', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='blue')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
    

plot_df(ap, x=ap['Month'], y=ap['Number of Passengers'], title='Trend and Seasonality')
```
